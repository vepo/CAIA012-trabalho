```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
GF = nx.read_gml("users-friends.gml")
print("Número de usuários: {}".format(len(GF.nodes())))
print("Número de amizades: {}".format(len(GF.edges())))
print("Número de médio de amigos: {}".format(len(GF.edges()) / len(GF.nodes())))
```

```python
GS = nx.read_gml("users-similarity.gml")
print("Número de usuários: {}".format(len(GS.nodes())))
print("Número de similaridades: {}".format(len(GS.edges())))
print("Número de médio de similares: {}".format(len(GS.edges()) / len(GS.nodes())))
```

```python
friend_similarity = []
for edge in GF.edges():
    if GS.has_edge(edge[0], edge[1]):
        friend_similarity.append(GS.edges()[edge]['weight'])
    else:
        friend_similarity.append(0)

plt.hist(friend_similarity, bins=51)
plt.ylabel('Quantidade')
plt.xlabel('Correlação');
```
